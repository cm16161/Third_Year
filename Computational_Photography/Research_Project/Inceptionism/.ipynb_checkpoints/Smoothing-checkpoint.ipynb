```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
from PIL import Image
```

```python
def line_enhancement(imgBase, imgBody, edgeMethod="laplacian", alpha = 0.1):
    if(edgeMethod == "laplacian"):
        edgeImg = cv2.Laplacian(imgBase,cv2.CV_64F)
    if(edgeMethod == "sobelX"):
        edgeImg = cv2.Sobel(imgBase,cv2.CV_64F,1,0,ksize=5)
    else:#(edgeMethod == "sobelY"):
        edgeImg = cv2.Sobel(imgBase,cv2.CV_64F,0,1,ksize=5)

    height, width = imgBase.shape[:2]
    imgBody =cv2.resize(imgBody,(width,height))

    beta=1-alpha
    gamma=0
    dst = np.uint8(alpha*(edgeImg)+beta*(imgBody))
    dst = cv2.applyColorMap(dst,cv2.COLORMAP_JET)
    plt.subplot(121), plt.imshow(dst)
    return dst
```

```python
img = cv2.imread("Strong-Lines-San-Francisco.jpg",0) 
img2 = cv2.imread("flowers.jpg",0)
blur = cv2.blur(img,(10,10),0)
output = line_enhancement(blur,img2,"sobelY")
plt.imshow(output)
cv2.imwrite("testBlur.png",output)
```
